**Visualizing in Python Using seaborn**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
irisData = sns.load_dataset('iris')
print(irisData)
```

```python
irisData.describe()
```

```python
grpdData = irisData.groupby('species')
grpdData.count()
```

```python
irisDataSetosa = irisData.loc[irisData.species == 'setosa',
['petal_length','petal_width']]

irisDataVersicolor = irisData.loc[irisData.species == 'versicolor',
['petal_length','petal_width']]

irisDataVirginica = irisData.loc[irisData.species == 'virginica',
['petal_length','petal_width']]
```

#### One thing to understand about Matplotlib is that, it doesn't work well with pandas dataframe that is the reason we created pandas series from dataframe and plot individually each series, one at a time. 


```python
fig = plt.figure(figsize=(12, 4))

# 1 row , 3 column, 1st plot 
subp1 = fig.add_subplot(131)
subp1.scatter(irisDataSetosa.petal_length, irisDataSetosa.petal_width,marker= 'o',color='olive')
subp1.set_title('Setosa')
subp1.set_xlabel('Petal Length')
subp1.set_ylabel('Petal width')

# 1 row , 3 column, 2nd plot 
subp2 = fig.add_subplot(132)
subp2.scatter(irisDataVersicolor.petal_length,irisDataVersicolor.petal_width,color='r')
subp2.set_title('Versicolor')
subp2.set_xlabel('Petal Length')
subp2.set_ylabel('Petal width')

# 1 row , 3 column, 2nd plot 
subp3 = fig.add_subplot(133)
subp3.scatter(irisDataVirginica.petal_length,
irisDataVirginica.petal_width)
subp3.set_title('Virginica')
subp3.set_xlabel('Petal Length')
subp3.set_ylabel('Petal width')
;
```

```python
sns.boxplot(x='species',y='petal_width',data=irisData);
```

```python
sns.boxplot(x='species',y='petal_length',data=irisData);
```

```python
sns.pairplot(data=irisData,x_vars=['petal_length','petal_width'],y_vars=['petal_length','petal_width']);
```

Following along the default tips data set of the seaborn tutorial from the seaborn library

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(3)
```

```python
tips.describe()
```

```python
print(tips.day.unique())
print(tips.time.unique())
```

#### lmplot plot a liner model on the data

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips);
```

```python
sns.lmplot(x="total_bill", y="tip",
data=tips,
height=3.5,
aspect=2);
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue='smoker',  markers=["o", "x"]);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
               palette="Set1");
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips);
```

```python
sns.set(font_scale=2) 
sns.lmplot(x="total_bill", y="tip", col="day", data=tips);
```

```python
plt.figure(figsize=(6,3.5))
sns.set(font_scale=1) 
sns.boxplot(x='smoker',y ='tip',data = tips);
```

```python
plt.figure(figsize=(10,3.5))
sns.set(font_scale=1) 
sns.set_style('dark') 
sns.boxplot(x='day',y ='tip',data = tips, hue = 'smoker');
```

#### Stripe plot and Swarm Plot

```python
sns.stripplot(x='sex',y ='tip',data = tips);
plt.grid(True)
```

```python
sns.swarmplot(x='sex',y ='total_bill',data = tips);
plt.grid(True)
```

```python
sns.countplot(x='sex',data = tips);
```

```python
sns.jointplot('tip','total_bill',data = tips, kind="kde",height = 4);
```

```python
sns.jointplot('tip','total_bill',data = tips, kind="hex",height = 4);
```
